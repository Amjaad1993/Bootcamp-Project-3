### Import your librares

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### Loading Data

```python
df = pd.read_csv('RiyadhVillasAqar.csv')
pd.set_option('display.max_columns', None)
df
```

## EDA

```python
print(df.head())
nan_count = df.isna().sum().sum()
print('NaN Values = ',nan_count)
print('Shipe is ',df.shape)
print(df.dtypes)
Print(df.duplicated)
```

## 10 Most Expensive Neighborhoods

```python
df1 = df.copy()
mean = df1.groupby('neighbourhood')['square price'].mean()

mean.index
df2 = mean.reset_index()
df2['square price'] = df2['square price'].round(decimals=2)
df2 = df2.sort_values(by='square price',ascending=False).head(10)

fig = px.bar(df2, x='neighbourhood', y='square price',color='neighbourhood',title='10 Most Expensive Neighborhoods',color_discrete_sequence=px.colors.qualitative.Vivid, labels={'square price':'Average Square Price', 'neighbourhood':'Neighbourhood'})
fig.show()
```

## Average Square Price VS Property Age

```python
df1 = df.copy()
mean = df1.groupby('propertyAge')['square price'].mean()
df2 = mean.reset_index()
df2['square price'] = df2['square price'].round(decimals=1)

fig = px.line(df2, x='propertyAge', y='square price', markers=True,title='Average Square Price VS Property Age',color_discrete_sequence=px.colors.qualitative.Vivid,labels={'square price':'Average Square Price', 'propertyAge':'Property Age'})
fig.show()
```

## Available Properties

```python

df1 = df.copy()

fig = px.scatter(df1, x="propertyAge", y="price",
        size="space", color="location", hover_name="location",size_max=45,color_discrete_sequence=px.colors.qualitative.Vivid, labels={'location':'Location', 'price':'Price','propertyAge':'Property Age'},title='Available Properties')
fig.show()
```
