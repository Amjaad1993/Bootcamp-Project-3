```python
# Filter rows where space is larger than 6000 sqft
d1 = df[df['space'] > 6000]

# Group data by location and calculate the mean of the price
d1 = d1.groupby('location').mean()

# Create a line chart to visualize the data
sns.lineplot(data=d1, x=d1.index, y='price')

# Add title and labels
plt.title("Average price of villas with space > 6000")
plt.xlabel("location")
plt.ylabel("price")
# Show the chart
plt.show()
```

```python
# Create a bar chart to visualize the data
sns.countplot(x='location',hue='location', data=df)

# Add title and labels
plt.title("distribution of villas by location")
plt.xlabel("location")
plt.ylabel("number of villas")

# Show the chart
#place legend in center right of plot
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, title='location')
plt.show()
```

```python
sns.scatterplot(x='space', y='price', data=df)
plt.title("relationship between space and price")
plt.show()
```
