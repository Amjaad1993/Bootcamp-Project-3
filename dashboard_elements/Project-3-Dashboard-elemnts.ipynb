## Handling Missing Data

```python
df.lounges.isnull().sum()
```

```python
# To handle the plus sign with 7+
df['lounges'] = df['lounges'].str.replace('+','')

print(df['lounges'].isnull().sum())
df['lounges'] = df['lounges'].apply(pd.to_numeric)
# Take the mean for lounges missing data 
lounges_mean = round(df['lounges'].mean())
df['lounges'] = df['lounges'].fillna(lounges_mean)
print(df['lounges'].isnull().sum())
```

```python
# Take the mean for streetWidth missing data

streetWidth_mean = round(df['streetWidth'].mean())
df['streetWidth'] = df['streetWidth'].fillna(streetWidth_mean)
df['streetWidth'].isnull().sum()
```

```python
# Take the mean for price missing data

price_mean = round(df['price'].mean())
df['price'] = df['price'].fillna(price_mean)
df['price'].isnull().sum()
```

```python
# Take the mean for price missing data

squareprice_mean = round(df['square price'].mean())
df['square price'] = df['square price'].fillna(squareprice_mean)
df['square price'].isnull().sum()
```

# Imports

```python
from jupyter_dash import JupyterDash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output
```

# Dashboard element shows the price distribution of each location

```python
app = JupyterDash()
app.layout = html.Div(children = [
    dcc.Dropdown(
        id='location-dropdown',
        options=[{'label': location, 'value': location} for location in df['location'].unique()],
        value='All'
    ),
    dcc.Graph('my-graph')
])
@app.callback(
Output('my-graph', 'figure'),
Input('location-dropdown','value')
)
def update_graph(location):
    fillter_df = df[df['location']==location ]
    histogram = px.histogram(fillter_df , x='price')
    
    return histogram


if __name__ == '__main__':
    app.run_server(debug=False,port=8050)
```

# Dashborad element shows the price vs space scatter plot of each neighbourhood

```python
app = JupyterDash()
app.layout = html.Div(children = [
    dcc.Dropdown(
        id='neighbourhood-dropdown',
        options=[{'label': neighbourhood, 'value': neighbourhood} for neighbourhood in df['neighbourhood'].unique()],
        value='All'
    ),
    dcc.Graph('my-graph')
])
@app.callback(
Output('my-graph', 'figure'),
Input('neighbourhood-dropdown','value')
)
def update_graph(neighbourhood):
    fillter_df = df[df['neighbourhood']==neighbourhood ]
    scatter = px.scatter(fillter_df , x='price',y='space')
    
    return scatter


if __name__ == '__main__':
    app.run_server(debug=False,port=8040)
```

# Dashborad element shows the price distribuation of for the chosen options

```python
app = JupyterDash()
app.layout = html.Div(children = [
    html.H2('Neighborhood'),
    dcc.Dropdown(
    id='neighbourhood-dropdown',
    options=[{'label': location, 'value': location} for location in df['neighbourhood'].unique()],
    value='All'
    ),
    
    html.H2('Number of rooms'),
    dcc.Dropdown(
        id='room-dropdown',
        options=[{'label': room, 'value': room} for room in df['rooms'].unique()],
        value='All'
    ),
    
    html.H2('Age of property'),
    dcc.Dropdown(
    id='age-dropdown',
    options=[{'label': age, 'value': age} for age in df['propertyAge'].unique()],
    value='All'
    ),
    
    dcc.Graph('my-graph')
])
@app.callback(
Output('my-graph', 'figure'),
[Input('neighbourhood-dropdown','value'),Input('room-dropdown','value'),Input('age-dropdown','value')]
)
def update_graph(neighbourhood,room,age):
    fillter_df = df[(df['rooms']==room) & (df['propertyAge']==age) & (df['neighbourhood']==neighbourhood)]
    histogram = px.histogram(fillter_df , x='price')
    return histogram


if __name__ == '__main__':
    app.run_server(debug=False,port=8030)
```
